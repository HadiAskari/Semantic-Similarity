```python
materialID=df['lastRunJob'].to_list()
```

```python
materialtitle=[]
seriestitle=[]
```

```python
for ID in tqdm(materialID):
    asset_metadata = json.loads(openUrl('http://ac3448e420fce11eaaa4b0a458c10dab-684955813.us-east-1.elb.amazonaws.com/MafData/rs/db/mafdb/metadata/{}'.format(ID)).read())
    materialtitle.append(asset_metadata['tags']['materialTitle'])
    seriestitle.append(asset_metadata['tags']['seriesTitle'])
    
```

```python
CC=[]
```

```python
asset_metadata = json.loads(openUrl('http://ac3448e420fce11eaaa4b0a458c10dab-684955813.us-east-1.elb.amazonaws.com/MafData/rs/db/mafdb/analysis/{}/ClosedCaptions?limit=100000'.format(ID)).read())
```

```python
for ID in tqdm(materialID):
    cc1=[]
    try:
        asset_metadata = json.loads(openUrl('http://ac3448e420fce11eaaa4b0a458c10dab-684955813.us-east-1.elb.amazonaws.com/MafData/rs/db/mafdb/analysis/{}/ClosedCaptions?limit=100000'.format(ID)).read())
        for i in range(len(asset_metadata['result'])):
            temp=asset_metadata['result'][i]['text']
            temp=temp.replace('-','')
            cc1.append(temp)
        string=' '.join([str(item) for item in cc1])
        CC.append(string)
    except:
        CC.append('-')
```

```python
df['materialtitle']=materialtitle
df['seriestitle']=seriestitle
df['AssetCC']=CC
```
